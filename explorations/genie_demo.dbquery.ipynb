```sql

USE CATALOG marathos;
USE SCHEMA gold;


-- ============================================================
-- Q1: Top 10 countries by number of finishers
-- ============================================================
SELECT
  c.country_name,
  c.continent,
  COUNT(*) AS finishers
FROM fct_results f
JOIN dim_country c USING (country_id)
GROUP BY c.country_name, c.continent
ORDER BY finishers DESC
LIMIT 10;


-- ============================================================
-- Q2: Average finish time for 100km races (in hours)
-- ============================================================
SELECT
  COUNT(*) AS finishers,
  ROUND(AVG(f.performance_seconds) / 3600.0, 2) AS avg_hours,
  ROUND(MIN(f.performance_seconds) / 3600.0, 2) AS fastest_hours
FROM fct_results f
JOIN dim_event e USING (event_id)
WHERE e.event_type = 'distance'
  AND e.event_value = 100
  AND e.event_unit = 'km';


-- ============================================================
-- Q3: Number of finishers per year
-- ============================================================
SELECT
  d.year,
  COUNT(*) AS finishers
FROM fct_results f
JOIN dim_date d USING (date_id)
GROUP BY d.year
ORDER BY d.year;


-- ============================================================
-- Q4: Top 10 athletes by distance covered in 24-hour races
-- ============================================================
SELECT
  a.athlete_id,
  a.athlete_gender,
  c.country_code,
  e.event_name,
  f.performance_km,
  d.year
FROM fct_results f
JOIN dim_event   e USING (event_id)
JOIN dim_athlete a USING (athlete_id)
JOIN dim_country c USING (country_id)
JOIN dim_date    d USING (date_id)
WHERE e.event_type = 'duration'
  AND e.event_value = 24
ORDER BY f.performance_km DESC
LIMIT 10;


-- ============================================================
-- Q5: Female vs male average performance in 50km races
-- ============================================================
SELECT
  a.athlete_gender,
  COUNT(*) AS finishers,
  ROUND(AVG(f.performance_seconds) / 3600.0, 2) AS avg_hours,
  ROUND(AVG(f.athlete_avg_speed_kmh), 2)        AS avg_speed_kmh
FROM fct_results f
JOIN dim_event   e USING (event_id)
JOIN dim_athlete a USING (athlete_id)
WHERE e.event_type = 'distance'
  AND e.event_value = 50
  AND e.event_unit = 'km'
  AND a.athlete_gender IN ('M', 'W')
GROUP BY a.athlete_gender
ORDER BY a.athlete_gender;
```
